```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics as sts
import seaborn as sns
```

```python
student_name = ['Piter Parker', 'Rick Sanchez', 'Bruce Batman', 'Vitaliy Buterin', 'Mikasa Akkerman', 'Robert House']
devess_attendance = np.array([3/18*100, 0/18*100, 7/18*100, 18/18*100, 18/18*100, 15/18*100])
grade = np.array([85, 99, 83, 97, 100, 100])

df = pd.DataFrame(data={'student_name':student_name,
                        'devess_attendance':devess_attendance,
                        'grade': grade})

df
```

```python
plt.subplots()

sns.heatmap( df.corr(), cmap='coolwarm', annot=True)
plt.title("Correlation matrix - attendance and grades")

sns.lmplot(x='devess_attendance', y='grade', data=df)
plt.title("Correlation between attendance and grades")
plt.show()
```
